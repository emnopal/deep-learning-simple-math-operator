```python
import sys
sys.path.append('..')

import matplotlib.pyplot as plt
import numpy as np

from dataset.generator import generate_dataset
from model.valid_load import load_model, predicting_model, print_accuracy
```

```python
length_default_dataset_with_zeros = generate_dataset(
    operation='add', num=100, include_zeros=True)[0].shape[0]
length_default_dataset_without_zeros = generate_dataset(
    operation='add', num=100, include_zeros=True)[0].shape[0]
```

```python
n = 1000
X, y = generate_dataset(operation='add', num=n, include_zeros=True)

model = load_model('../saved_model/addition.h5')
pred = predicting_model(model, X, ndim=1)

results = np.sort(y.values)
predicted = np.sort(pred)
length = range(len(y))

print(print_accuracy(results, predicted))

plt.plot(results[:length_default_dataset_with_zeros],
         length[:length_default_dataset_with_zeros], 'y')
plt.plot(predicted[:length_default_dataset_with_zeros],
         length[:length_default_dataset_with_zeros], 'r')
plt.plot(results[length_default_dataset_with_zeros:, ],
         length[length_default_dataset_with_zeros:], 'b--', linewidth=2)
plt.plot(predicted[length_default_dataset_with_zeros:],
         length[length_default_dataset_with_zeros:], 'g--', linewidth=2)

plt.legend(['Interpolation Real', 'Interpolation Predicted',
           'Extrapolation Real', 'Extrapolation Predicted'], bbox_to_anchor=(1, 1))

plt.show()
```
